```python
import yfinance as yf
import pandas as pd

```

```python
tickers = [ "MSFT", "AAPL", "NVDA", "AMZN", "GOOGL", "META", "TSLA"]

# The data is pulled daily.
data = yf.download(tickers, start="2014-01-01", end="2023-12-31")

# Converting daily data to yearly data.
yearly_data = data.resample('Y').last()

```

```python
# View the data.
print(yearly_data)
```

```python
# Display the first 5 data points.
yearly_data.head()
```

```python
yearly_data.describe()
```

```python
yearly_data.info()
```

```python
# We will proceed with the implementation based on the high value of the stocks.
df = yearly_data['High']
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# We have set the 'Date' column as the index.
df.index = df.index.strftime ('%Y-%m-%d')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
# We are examining the changes in stock prices through a graph
df.plot()
```

```python
price_increase = df / df.iloc[0] * 100

# Create a graph.
plt.figure(figsize=(10, 6))

# Plot the increase rates for each stock.
for ticker in price_increase.columns:
    plt.plot(price_increase.index, price_increase[ticker], label=ticker)

# Define the graph properties.
plt.title('Hisse Senedi Yüksek Değerlerinin Zamana Göre Artışı')
plt.xlabel('Tarih')
plt.ylabel('Başlangıç Fiyatına Göre Artış Oranı (%)')
plt.legend(loc='upper left')
plt.grid(True)
```
